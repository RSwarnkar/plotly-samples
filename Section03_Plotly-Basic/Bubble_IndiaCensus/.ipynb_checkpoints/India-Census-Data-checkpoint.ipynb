```python
import pandas as pd
import plotly.graph_objs as go
import plotly.offline as pyo
import numpy as np
```

```python
df = pd.read_csv("India-census-Data.csv", index_col='States')
```

```python
df.head()
```

```python
df.columns
```

```python
import matplotlib
%matplotlib inline

df.sort_values('CrimeIncidents-Total',inplace=True)
```

```python
df['CrimeIncidents-Total'].plot(kind='bar',sort_columns=True,logy=True,figsize=(12,8),colormap="coolwarm")
```

```python
df['CrimeIncidents-Total']

df['log_CrimeIncidents-Total'] = np.log(df['CrimeIncidents-Total']) 
df.head()
```

```python
#df.dropna(subset = ['column1_name', 'column2_name', 'column3_name'])
df.dropna(inplace=True)
df.head()
```

# Bubble Plot will be a 4D data with below:
* x-axis = Population
* y-axis = Per Capita Income
* bubble radius size = Crime rate
* Bubbles = individual states




```python
mydata = [go.Scatter(x=df['Population'],
                     y=df['Net Domestic Product per capita (INR)'],
                     text=df.index,
                     mode='markers',
                     marker=dict(
                         size=3*df['log_CrimeIncidents-Total'],
                         color=df['CrimeIncidents-Total'],
                         showscale=True
                     )
                    )]

mylayout = go.Layout(title="Indian States Census Data",
                    xaxis=dict(title='Population'),
                    yaxis=dict(title='Net Domestic Product per capita (INR)'.),
                     hovermode="closest"
                    )

myfig = go.Figure(data=mydata, layout=mylayout)
```

```python
pyo.plot(myfig,filename="India-Census01.html")
```
